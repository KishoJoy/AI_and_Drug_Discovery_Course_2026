Target Selection and Data Retrieval

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client

target = new_client.target
target_query = target.search('ERBB2')
targets = pd.DataFrame.from_dict(target_query)

selected_target = 'CHEMBL1824'
print(f"selected Target ChEMBL ID: {selected_target}")

activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
df = pd.DataFrame.from_dict(res)

df = pd.DataFrame.from_dict(res)

df.to_csv('bioactivity_data_raw.csv', index=False)
print(f"Total raw records retrieved: {len(df)}")
```

Data Prpcessing and Classification

```python
df2 = df.dropna(subset=['standard_value', 'canonical_smiles'])

df2 = df2[df2.standard_units == 'nM']

bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")

selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df_final = df2[selection].copy()
df_final['class'] = bioactivity_class

df_final.to_csv('Assignment_2_QSAR_curated_data.csv', index=False)
print("Curated data saved successfully!")
df_final.head()
```
